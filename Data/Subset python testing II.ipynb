```python
weather = pd.read_csv('./Downloads/weather.csv')
```

```python
weather_short = pd.melt(weather,id_vars=['id', 'year', 'month', 'element'],
    var_name='day',
    value_name='temp')
```

```python
weather_short.head()
```

```python
weather_tidy = (weather_short.pivot_table(
index = ['id', 'year', 'month', 'day'],
columns = 'element',
values = 'temp')
               .reset_index()
               )
```

```python
ebola = pd.read_csv('./Downloads/country_ts.csv')
```

```python
ebola.head()
```

```python
ebola_long = pd.melt(ebola, id_vars = ['Date', 'Day'])
ebola_long.head()
```

```python
variable_split = ebola_long ['variable'].str.split('_')
variable_split.head()
```

```python
status_values = variable_split.str.get(0)
status_values.head()
```

```python
country_values = variable_split.str.get(0)
country_values.head()
```

```python
ebola_long['status'] = status_values
ebola_long['country'] = country_values
```

```python
ebola_long.head(20)
```

```python
ebola_long = pd.concat([ebola_long, variable_split], axis = 1)
```
